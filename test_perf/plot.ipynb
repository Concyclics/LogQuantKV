```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('speed.csv')
df['len'] = df['output_len'] // df['batch_size']
df
```

```python
df_1000 = df[df['len'] == 1000]
df_2000 = df[df['len'] == 2000]
df_1000
```

```python
df_1000 = df_1000[df_1000['memory'] <= 48]
df_1000
```

```python
df_2000 = df_2000[df_2000['memory'] <= 48]
df_2000
```

```python
sns.lineplot(x='batch_size', y='speed', data=df_2000, hue='method', markers=True, style='method')
```

```python
df_2000['method'] = df_2000['method'].replace({'baseline': 'Baseline(BF16)', 'LogQuant': 'LogQuant(2-bit)'})
```

```python
sns.set_theme(style='whitegrid', font_scale=1.5)
fig, ax = plt.subplots(1, 3, figsize=(16, 5))
#修改所有字体大小
#sns.set_context("talk")
#修改坐标轴上tixk字体大小

sns.lineplot(x='batch_size', y='speed', data=df_2000, hue='method', markers=True, style='method', palette='tab10', ax=ax[0], legend=False, linewidth=2, markersize=8)
ax[0].set_xlabel('Batch Size', fontsize=20)
ax[0].set_ylabel('Speed (Tokens/s)', fontsize=20)
sns.lineplot(x='memory', y='speed', data=df_2000, hue='method', markers=True, style='method', palette='tab10', ax=ax[1], legend=False, linewidth=2, markersize=8)
ax[1].set_xlabel('Memory (GB)', fontsize=20)
ax[1].set_ylabel('Speed (Tokens/s)', fontsize=20)
sns.lineplot(x='memory', y='batch_size', data=df_2000, hue='method', markers=True, style='method', palette='tab10', ax=ax[2], legend=True, linewidth=2, markersize=8)
ax[2].set_xlabel('Memory (GB)', fontsize=20)
ax[2].set_ylabel('Batch Size', fontsize=20)

plt.legend(ncol = 2, bbox_to_anchor=(-0.069, -0.25), fontsize=20)
#调整subplot之间的距离
plt.subplots_adjust(top=0.9, bottom=0.2, left=0.1, right=0.9, hspace=0.5, wspace=0.3)
plt.tight_layout()
plt.savefig('speed_vs_memory.pdf', dpi=600, bbox_inches='tight')
```
