```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('speed.csv')
df['len'] = df['output_len'] // df['batch_size']
df
```

```python
df_1000 = df[df['len'] == 1000]
df_2000 = df[df['len'] == 2000]
df_1000
```

```python
df_1000 = df_1000[df_1000['memory'] <= 48]
df_1000
```

```python
df_2000 = df_2000[df_2000['memory'] <= 48]
df_2000
```

```python
sns.lineplot(x='batch_size', y='speed', data=df_2000, hue='method', markers=True, style='method')
```

```python
sns.lineplot(x='memory', y='speed', data=df_2000, hue='method', markers=True, style='method', dashes=False)
```

```python
sns.lineplot(x='batch_size', y='memory', data=df_2000, hue='method', markers=True, style='method', dashes=False)
```
